```python
import os
import json
import pandas as pd
```

```python
# Define the root directory for results
results_dir = "./results"

# Define the keys to extract from each .result file
keys_to_extract = [
    "n_samples", 
    "algorithm", 
    "ddim.schedule_params.jump_length", 
    "ddim.schedule_params.jump_n_sample", 
    # "ddim.schedule_params.use_timetravel", 
    "optimize_xt.num_iteration_optimize_xt", 
    "optimize_xt.lr_xt",
    "lpips_score_sample0",
    "psnr_score_sample0",
    "ssim_score_sample0",
    "mean time"
]

# Initialize a list to store data from all files
all_results = []

# Loop through the directory structure
for folder in os.listdir(results_dir):
    current_folder = os.path.join(results_dir, folder)
    for file in os.listdir(current_folder):
        if file.endswith(".result"):  # Check if the file is a `.result` file
            file_path = os.path.join(current_folder, file)
            
            # Extract folder name for context
            folder_name = folder
            
            # Open and read the .result file
            with open(file_path, 'r') as f:
                data = f.readlines()
            
            # Parse the .result file
            extracted_data = {}
            for line in data:
                try:
                    entry = json.loads(line.strip())  # Parse each line as JSON
                    # Extract only the relevant keys
                    for key in keys_to_extract:
                        if key in entry:
                            extracted_data[key] = entry[key]
                except json.JSONDecodeError:
                    print(f"Skipping invalid line in {file_path}: {line}")
            
            # # Add folder name as a context column
            # extracted_data['folder_name'] = folder_name
            
            # Append to the results list
            all_results.append(extracted_data)
```

```python
# Create a DataFrame from the results
df = pd.DataFrame(all_results)
print(df.shape)
df.head()
```

```python
df.to_csv('./results.csv', index=False)
```
